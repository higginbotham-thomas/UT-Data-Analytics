```python
import numpy as np
import pandas as pd
import pandas_profiling
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn.datasets
from sklearn.tree import plot_tree
%matplotlib inline
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import export_graphviz
```

```python
#Loading data and checking the top 5 rows
df = pd.read_csv('Demographic_Data.csv')
df.head()
```

```python
#Info on that data set
df.info()
```

```python
#Preprocessing - creating decade buckets that will be used to cut the ages
decades = ([1, 9, 19, 29, 39, 49, 59, 69, 79, 89])
decades
```

```python
#Preprocessing - creating amount buckets that will be used to cut the amounts
amount_bucket = (0, 1000, 2000, 3000)
amount_bucket
```

```python
#Discretizing age into decades and providing information on the dataset after adding the new discretized "age_label" column
df['age_label'] = pd.cut(df['age'], decades, labels = False)
df.info()
```

```python
#Discretizing amounts into buckets and providing information on the dataset after adding the new discretized "amount_label" column
df['amount_label'] = pd.cut(df['amount'], amount_bucket, labels = False)
df.info()
```

```python
#First 10 rows to confirm that the new labels are working properly
df.head(10)
```

```python
#Dropping the original age and amount columns for use moving forward
df = df.drop(columns = ['age'])
df = df.drop(columns = ['amount'])
```

```python
#Confirming that these have been dropped properly by looking at the first 5 rows
df.head()
```

```python
#Modeling using the three algorithms - Decision Tree Classifier, Random Forest Classifier, and Gradient Boosting Classifier
algos_class = []
algos_class.append(('Decision Tree Classifier', DecisionTreeClassifier()))
algos_class.append(('Random Forest Classifier', RandomForestClassifier()))
algos_class.append(('GradientBoostingClassifier', GradientBoostingClassifier()))
```

```python
#Setting the discretized age (decades) as the dependent variable
x = df[['region', 'items', 'amount_label', 'in-store']]
print(x.head())

y = df['age_label']
print(y.head())
```

```python
#Train and test for age dependent variable
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .30, random_state = 123)

print('Data shape', x_train.shape, y_train.shape)
```

```python
#Classification - creating empty lists to store results and names of the algorithm for printing
#Setting cross validation folds of 3
results = []
names = []

for name, model in algos_class:
    result = cross_val_score (model, x_train, y_train, cv=3, scoring ='accuracy')
    names.append(name)
    results.append(result)
    
for i in range(len(names)):
    print(names[i], results[i].mean())
```

```python
#Printing the results and names
#The three algorithms had similar scores so will utilize the Decision Tree Classifier for ease of use
print(results)
print(names)
```

```python
#Modeling classification for discretized age
algo = DecisionTreeClassifier(max_depth=3)
dtc = algo.fit(x_train, y_train)

#Prediction and report
#Accuracy was very low for predicting age through this - will not continue plotting graph
preds = dtc.predict(x_test)
print(classification_report(y_test, preds))
```

```python
#Dependent variable of Amount
x = df[['region', 'items', 'in-store', 'age_label']]
print(x.head())

y = df['amount_label']
print(y.head())
```

```python
#Train and Test split for discretized amount as dependent variable along with shape of train data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .30, random_state = 123)

print('Data shape', x_train.shape, y_train.shape)
```

```python
#Classification
results = []
names = []

for name, model in algos_class:
    result = cross_val_score (model, x_test, y_test, cv=3, scoring ='accuracy')
    names.append(name)
    results.append(result)
    
for i in range(len(names)):
    print(names[i], results[i].mean())
```

```python
#Printing the results and names
#The three algorithms had similar scores so will utilize the Decision Tree Classifier for ease of use
print(results)
print(names)
```

```python
#Modeling classification for discretized amount
algo = DecisionTreeClassifier(max_depth=3)
dtc1 = algo.fit(x_train, y_train)

#Prediction and report
#Accuracy of prediction was 0.71 so will plot out decision tree for insight
preds = dtc1.predict(x_test)
print(classification_report(y_test, preds))
```

```python
#Plotting decision tree
plt.figure(figsize=(25,10))
plot_tree(dtc1,
         feature_names = x.columns,
         class_names =['online','in-store'],
          filled = True,
          fontsize = 13,
          max_depth = 3)
plt.show()
```

```python
#Dependent variable of In-store
x = df[['region', 'items', 'amount_label', 'age_label']]
print(x.head())

y = df['in-store']
print(y.head())
```

```python
#Train and test for in-store dependent variable
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .30, random_state = 123)

print('Data shape', x_train.shape, y_train.shape)
```

```python
#Classification
results = []
names = []

for name, model in algos_class:
    result = cross_val_score (model, x_test, y_test, cv=3, scoring ='accuracy')
    names.append(name)
    results.append(result)
    
for i in range(len(names)):
    print(names[i], results[i].mean())
```

```python
#Printing the results and names
#The three algorithms had similar scores so will utilize the Decision Tree Classifier for ease of use
print(results)
print(names)
```

```python
#Modeling classification for in-store purchase
algo = DecisionTreeClassifier(max_depth=3)
dtc2 = algo.fit(x_train, y_train)

#Prediction and report
#Accuracy was 0.84 so will proceed to plotting
preds = dtc2.predict(x_test)
print(classification_report(y_test, preds))
```

```python
#Plotting decision tree of in-store as dependent variable - All in-store 
#Region 1 shopped exclusively in-store - color evidence with dark blue majority
plt.figure(figsize=(25,10))
plot_tree(dtc2,
         feature_names = x.columns,
         class_names =['online','in-store'],
          filled = True,
          fontsize = 13,
          max_depth = 3)
plt.show()
```

```python
#Loading the original dataset for correlations
#Slight correlation between age and online/in-store purchase 
#Age was the strongest correlation for all variables in relation to in-store purchase
#Stronger correlation between age and amount and even stronger correlation between age and region
demodata = pd.read_csv('Demographic_Data.csv')
corr_mat = demodata.corr()
print(corr_mat)
```

```python
#Covariance
cov_mat = demodata.cov()
print(cov_mat)
```

```python
#Plotting average amount purchased by age
#Illustrates that as age increases, amount does not increase
plt.figure(figsize=(25,10))
Average_amount = demodata.groupby('age')['amount'].mean()
Average_amount.plot(kind='bar', title="Average Purchase amount by Age", ylabel = 'Average Amount Purchased', xlabel='age')
```

```python
#Average Age by Region
Average_age = demodata.groupby('region')['age'].mean()
Average_age.plot(kind='bar', title="Average Consumer Age by Region", ylabel='Average Age', xlabel='Region')
```

```python
#Average age of in-store or online purchase
#Highlights that most shoppers online are older customers
Average_age = demodata.groupby('in-store')['age'].mean()
print(Average_age)
```
