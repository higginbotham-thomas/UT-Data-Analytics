```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
#Reading the dataset
df = pd.read_csv('Demographic_Data.csv')
```

```python
#Dataset
df
```

```python
#Display first 5 rows
df.head()
```

```python
#Info on the data set
df.info()
```

```python
#Description of data
df.describe()
```

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
#Display the number of entries for each region
df['region'].value_counts()
```

```python
#Correlation
df.corr()
```

```python
#Covariance
df.cov()
```

```python
#Plotting Transaction Counts for each Region
plt.title("Transaction counts for each region")
sns_plot=sns.countplot(data=df,x='region')
fig=sns_plot.get_figure()
```

```python
#Plotting total counts by age
plt.figure(figsize=(20,5))
plt.title("Total counts by age")
sns_plot=sns.countplot(data=df,x='age')
fig=sns_plot.get_figure()
```

```python
#Plotting total count for number of items
plt.title("Total counts for items purchased per transaction")
sns_plot=sns.countplot(data=df,x='items')
fig=sns_plot.get_figure()
```

```python
#Boxplot for each data category
df.boxplot()
```

```python
#Histogram for age
df['age'].hist(bins=20)
```

```python
#Histogram for amounts
df['amount'].hist(bins=10)
```

```python
#Boxplot to illustrate distribution of amount
sns.catplot(x= 'amount', kind = 'box', data= df)
```

```python
#Scatter plot of age and amount, based on in-store
sns.relplot(x= 'age', y= 'amount', hue='in-store', data= df)
```

```python
#Scatter plot of region and amount, based on in-store
sns.relplot(x= 'region', y= 'amount', hue='in-store', data= df)
```

```python
#Linear plot of region against amount
sns.lmplot('region', 'amount', data= df)
```

```python
#Total amounts sold per region and shown in millions
Amount_Sold = df.groupby('region')['amount'].sum()
Amount_Sold/1e6
```

```python
#Grouping by region to show total amounts sold per region 
Amount_sold= df.groupby['region']['amount'].sum()
Amount_sold.plot(title= 'Total Amounts by Region', kind= 'bar', ylabel= 'Total amounts sold in tens of millions',
                xlabel= 'Region')
```

```python
#Average age from sales to determine potential in targeted marketing
Average_age = df.groupby('region')['age'].mean()
Average_age
```

```python
#Average Age by Region
Average_age = df.groupby('region')['age'].mean()
Average_age.plot(kind='bar', title="Average Consumer Age by Region", ylabel='Average Age', xlabel='Region')
```

```python
#Average Purchase Amount by Age
Average_amount = df.groupby('age')['amount'].mean()
Average_amount
```

```python
#Average Purchase Amount by Age
Average_amount = df.groupby('age')['amount'].mean()
Average_amount.plot(kind='bar', title="Average Purchase Amount by Age", ylabel = 'Average Amount Purchased', xlabel='Age',  figsize=(20, 5))
```

```python
#Average total transaction amount grouped by number of items in the transaction
Average_items = df.groupby('items')['amount'].mean()
Average_items
```
