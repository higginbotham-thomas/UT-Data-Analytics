```python
#make the first row into column headers
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
```

```python
#rename PAY_0 column to PAY_1
df.rename(columns = {'PAY_0':'PAY_1'}, inplace = True)
```

```python
#remove duplicate headers
df = df.loc[df['LIMIT_BAL'] != 'LIMIT_BAL']
```

```python
#drop any duplicate records
df = df.drop_duplicates()
```

```python
#replace spaces in column names
df.columns = df.columns.str.replace(' ', '_')
```

```python
#factorize sex column
df.loc[df['SEX'] == 'male', 'SEX'] = 0
df.loc[df['SEX'] == 'female', 'SEX'] = 1

#factorize education column
df.loc[df['EDUCATION'] == 'high school', 'EDUCATION'] = 0
df.loc[df['EDUCATION'] == 'university', 'EDUCATION'] = 1
df.loc[df['EDUCATION'] == 'graduate school', 'EDUCATION'] = 2
df.loc[df['EDUCATION'] == 'other', 'EDUCATION'] = 3

#factorize default_payment_next_month column
df.loc[df['default_payment_next_month'] == 'default', 'default_payment_next_month'] = 0
df.loc[df['default_payment_next_month'] == 'not default', 'default_payment_next_month'] = 1
```

```python
#unique values for a column
df.default_payment_next_month.unique()
```

```python
#convert all values in dataframe to int
df = df.astype (int)
```

```python
#find the number of months a payment was reported defaulted
df['DEFAULTED_NMOS']=df['PAY_1'].apply(lambda x: 1 if x >= 1 else 0)+df['PAY_2'].apply(lambda x: 1 if x >= 1 else 0)+df['PAY_3'].apply(lambda x: 1 if x >= 1 else 0)+df['PAY_4'].apply(lambda x: 1 if x >= 1 else 0)+df['PAY_5'].apply(lambda x: 1 if x >= 1 else 0)+df['PAY_6'].apply(lambda x: 1 if x >= 1 else 0)
```

```python
#find whether a default occurred AT ALL during the reporting periods
df['DEFAULTED'] = np.where(df.DEFAULTED_NMOS > 0, 1, 0)
```
